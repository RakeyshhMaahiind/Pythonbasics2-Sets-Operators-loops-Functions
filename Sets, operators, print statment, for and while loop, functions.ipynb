```python
# Union()

s1={1,2,3}
s2={3,4,5}

print(s1.union(s2))
print(s1|s2)
```

```python
#intersection

print(s1.intersection(s2))
print(s1&s2)
```

```python
# Difference
s1={1,2,3,4,5,6}
s2={5,6,7}
print(s1-s2)
print(s1.difference(s2))
print(s2-s1)
```

```python
#symmetric difference
s1={1,2,3,4,5}
s2={4,5,6,7,8}
print(s1.symmetric_difference(s2))
```

```python
#subset
s1={1,2,3,4,5}
s2={1,2,3}
print(s1.issubset(s2))
print(s2.issubset(s1))
```

```python
#intersection
s1={1,2,3,4,5}
s2={4,5,6,7,8}
s1.intersection_update(s2)
print(s1)
```

# Conditional Statements

```python
# if condition single condition
# condition to check weather num is positive

num=5

if num>0:
    print('{} is positive number'.format(num))
```

```python
num=1
if num>0:
    print('{} is positive number'.format(num))
```

```python
# if else condition

num=8

if num>0:
    print('positive')
else:
    print('Negative')
```

```python
# if elif else condition
num=0

if num>0:
    print('positive')
elif num<0:
    print('negative')
else:
    print('zero')
    
```

```python
# checking max value in 3 values
num1=10
num2=15
num3=20

if num1>num2 and num1>num3:
    print('number1 is biggest')
elif num2>num1 and num2>num3:
    print('number2 is biggest')
else:
    print('number3 is biggest')
```

```python
# Nested if condition
num=int(input("Enter some number:" ))
if num>=0:
    if num==0:
        print('zero number')
    else:
        print('positive number')
else:
    print('Negative number')
```

# For loop

```python
# for loop - looping or iterations, no conditions required
# for loop run until last element,last row or last item

snacks=['pizza','burger','apple']

for x in snacks:
    print(x)
```

```python
for x in 'python':
    print(x)
```

```python
for x in range(0,3):
    print(snacks [x])
```

```python
for x in range(15,20):
    print(x)
```

```python
# finding the odd number from 0 to 10
for x in range(0,11):
    if x%2==0:
        print(x, "is even")
    else:
        print(x, "is odd")
```

```python
# Nested for loop

list1=["google","facebook","yahoo"]
list2=[".com",".org",".in"]

for x in list1:
    for y in list2:
        print(x,y)
```

```python
list1=["google","facebook","yahoo"]
list2=[".com",".org",".in"]

for i in range(0,len(list1)):
    for j in range(0,3):
        print(i,j)
        print(list1[i],list2[j])
```

```python
# for loop with else
digit=[0,1,5]

for i in digit:
    print(i)
else:
    print('no items left')
```

# While Loop

```python
# while loop run until condition is true, once condition is false it will terminate the loop

num=10
while num<11:
    print('Hello')
    
# infinite loop
```

```python
# counter method for printing number 5 to 10

num=5
while num<=10:
    print(num)
    num=num+1
```

```python
num=5
sum1=0
while num<=10:
    print(num)
    sum1=sum1+num
    num=num+1
print('sum is',sum1)
```

```python
counter=0

while counter<3:
    print('inside end')
    counter= counter+1
else:
    print('inside else')
```

```python
# finding factorial of a value
num=int(input('enter some value'))
val=1

if num<0:
    print('no ! for negative values')
elif num==1 or num==0:
    print('1 is the factorial of {}'.format(num))
else:
    for x in range(1,num+1):
        val=val*x
print(val)
```
